```python
import pandas as pd
import os 
from yaml import safe_load
from munch import munchify
```

```python
path = "../data/rfc_2025d_cat.txt"
```

```python
def cs(a, b):
    return (a - 1, b)

colspecs = [
    cs(1, 14),     # Name (RFC source designator)
    cs(17, 24),    # Comnam (Common name)

    cs(27, 28),    # RAh (hours)
    cs(30, 31),    # RAm (minutes)
    cs(33, 41),    # RAs (seconds)

    cs(43, 43),    # DE- (sign)
    cs(44, 45),    # DEd (degrees)
    cs(47, 48),    # DEm (arcmin)
    cs(50, 57),    # DEs (arcsec)

    cs(60, 65),    # eRA (mas)
    cs(67, 72),    # eDE (mas)
    cs(76, 81),    # Corr

    cs(83, 89),    # Nobs
    cs(91, 96),    # Nsca
    cs(98, 102),   # Nses

    cs(105, 110),  # FsS
    cs(112, 117),  # FmS
    cs(119, 124),  # FlS

    cs(127, 132),  # FsC
    cs(134, 139),  # FmC
    cs(141, 146),  # FlC

    cs(149, 154),  # FsX
    cs(156, 161),  # FmX
    cs(163, 168),  # FlX

    cs(171, 176),  # FsU
    cs(178, 183),  # FmU
    cs(185, 190),  # FlU

    cs(193, 198),  # FsK
    cs(200, 205),  # FmK
    cs(207, 212),  # FlK

    cs(215, 222),  # MeaEpo
]

names = [
    "Name", "Comnam",
    "RAh", "RAm", "RAs",
    "DE_sign", "DEd", "DEm", "DEs",
    "eRA_mas", "eDE_mas", "Corr",
    "Nobs", "Nsca", "Nses",
    "FsS_Jy", "FmS_Jy", "FlS_Jy",
    "FsC_Jy", "FmC_Jy", "FlC_Jy",
    "FsX_Jy", "FmX_Jy", "FlX_Jy",
    "FsU_Jy", "FmU_Jy", "FlU_Jy",
    "FsK_Jy", "FmK_Jy", "FlK_Jy",
    "MeaEpo_yr",
]
```

```python
vlbi = pd.read_fwf(
    path,
    colspecs=colspecs,
    names=names,
    comment="#",
    header=None
)
vlbi.head(2)
```

```python
rfc_sources = vlbi.Name.apply(lambda x: x.split()[-1])
```

```python
with open(os.getenv("CONFIG_PATH")) as f:
    cfg = munchify(safe_load(f))

sources_w_maps = os.listdir(cfg.fits_path)
```

```python
sources_without_maps = list(set(rfc_sources).difference(set(sources_w_maps)))
print(sources_without_maps[:5], len(sources_without_maps), 21949 - 20498)
```
